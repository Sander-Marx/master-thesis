Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Fetch data into dataframe

```python
#Experiment 1
df = pd.read_csv("../Data/data_air_15000_liq_15000_complete.csv", sep=',')


print(df.info())
print("\nNumber of rows:", len(df))
print("Number of columns:", len(df.columns))

print("\nShape:", df.shape)
display(df.iloc[:, :])  # Show the table


print("\nSummary statistics:\n")
display(df.describe())

# Check missing values
print("\nMissing values:\n")
display(df.isnull().sum())

print("\nNumber of unique rows:", df.drop_duplicates().shape[0])
print("Number of duplicate rows:", len(df) - df.drop_duplicates().shape[0])
print("min a ", df['a'].min())
print("min b ", df['b'].min())

print("max a ", df['a'].max())
print("max b ", df['b'].max())
```

```python
# Distribution of alpha, theta and a,b parameters

# Filter data by target type
df_a = df[df["prefix"] == 'airfilled']
df_l = df[df["prefix"] == 'liquid']

df_a['a_mm'] = df_a['a'] * 1000
df_a['b_mm'] = df_a['b'] * 1000
df_l['a_mm'] = df_l['a'] * 1000
df_l['b_mm'] = df_l['b'] * 1000

fig, ((ax1, ax2),(ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(22, 22))

for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
    ax.tick_params(axis='both', labelsize=20)
    ax.xaxis.get_offset_text().set_fontsize(18)
    for spine in ax.spines.values():
        spine.set_linewidth(1.5)

#Color map
colors = {
    'incident_angle': '#674ea7',
    'alpha': '#8eacbb',
    'a_air': '#1a73e8',
    'b_air': '#1a73e8', 
    'a_liquid': 'tab:orange', 
    'b_liquid': 'tab:orange',
}


# 1. Distribution of Incident Angles
sns.histplot(data=df, x='theta_i_deg', bins=91, ax=ax1, color=colors['incident_angle'])
ax1.set_xticks(list(np.arange(0, 91, 10)))
ax1.set_title('Distribution of angle "θ" (Air- and Liquid-filled)', fontsize=24, pad=15)
ax1.set_xlabel('Angle (degrees)', fontsize=24)
ax1.set_ylabel('Count', fontsize=24)

# 2. Distribution of  alpha
sns.histplot(data=df, x='alpha', bins=16, ax=ax2, color=colors['alpha'])
ax2.set_title('Distribution of elongation "α" (Air- and Liquid-filled)', fontsize=24, pad=15)
ax2.set_xlabel('alpha value', fontsize=24)
ax2.set_ylabel('Count', fontsize=24)
ax2.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

# 3. Distribution of a for air-filled targets
sns.histplot(data=df_a, x='a_mm', bins=50, ax=ax3, color=colors['a_air'])
ax3.set_title('Distribution of semi-major axis "a" (Air-filled)', fontsize=24, pad=15)
ax3.set_xlabel('a value (mm)', fontsize=24)
ax3.set_ylabel('Count', fontsize=24)
ax3.ticklabel_format(style='plain', axis='x', scilimits=(0,0))

# 4. Distribution of b for air-filled targets
sns.histplot(data=df_a, x='b_mm', bins=50, ax=ax4, color=colors['b_air'])
ax4.set_title('Distribution of semi-minor axis "b" (Air-filled)', fontsize=24, pad=15)
ax4.set_xlabel('b value (mm)', fontsize=24)
ax4.set_ylabel('Count', fontsize=24)
ax4.ticklabel_format(style='plain', axis='x', scilimits=(0,0))

# 5. Distribution of a for liquid targets
sns.histplot(data=df_l, x='a_mm', bins=50, ax=ax5, color=colors['a_liquid'])
ax5.set_title('Distribution of semi-major axis "a" (Liquid-filled)', fontsize=24, pad=15)
ax5.set_xlabel('a value (mm)', fontsize=24)
ax5.set_ylabel('Count', fontsize=24)
ax5.ticklabel_format(style='plain', axis='x', scilimits=(0,0))

# 6. Distribution of b for liquid targets
sns.histplot(data=df_l, x='b_mm', bins=50, ax=ax6, color=colors['b_liquid'])
ax6.set_title('Distribution of semi-minor axis "b" (Liquid-filled)', fontsize=24, pad=15)
ax6.set_xlabel('b value (mm)', fontsize=24)
ax6.set_ylabel('Count', fontsize=24)
ax6.ticklabel_format(style='plain', axis='x', scilimits=(0,0))

# Adjust layout with more space at the top
plt.tight_layout(pad=3.0, rect=[0, 0, 1, 0.98])

# Add a main title for the entire figure with more space
fig.suptitle('Distribution of Parameters for Gas-filled and Liquid-filled Targets', 
             fontsize=24, y=0.99)

# Save figure
plt.savefig('../Figures/parameter_distributions.png', dpi=600, bbox_inches='tight')
plt.savefig('../Figures/parameter_distributions_low.png', dpi=100, bbox_inches='tight')
plt.show()
```
